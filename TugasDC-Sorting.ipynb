# Quick Sort

```python
# ALGORITMA QUICK SORT
def isiArray():
    # global A
    nilai_n = int(input("Nilai N: "))
    A = nilai_n *[None]
    for i in range(0,nilai_n,1):
        A[i] = int(input("Nilai x: "))
    print("Kondisi Awal Array:")
    printArray(A)
    print("Proses Quicksort")
    return A

def printArray(arr):
    print("", end="")
    for i in arr:
        if (i != None):
            print(i, end=" ")
    print()

def quickSort(leftMostIndex, rightMostIndex, A):
    '''
    Lakukan proses untuk membuat partisi
    Partisi kiri < pivot
    Partisi kanan > pivot
    '''
    pivot = A[rightMostIndex]
    left = leftMostIndex
    right = rightMostIndex - 1
    # right terbalik dengan left, Pivot harus pindah ke yang ditunjuk Left
    while(left<=right):
        # Cari nilai elemen > pivot
        while(A[left]<pivot):
            left = left + 1
        # Cari nilai elemen < pivot
        while(right>=leftMostIndex and A[right]>pivot):
            right = right - 1
        # Swap keduanya
        if left <= right:
            tmp = A[left]
            A[left] = A[right]
            A[right] = tmp
            left = left + 1
            right = right - 1
        printArray(A)
    # while sudah selesai
    # pindahkan pivot elemen ke posisi yang tepat
    tmp = A[left]
    A[left] = A[rightMostIndex]
    A[rightMostIndex] = tmp
    printArray(A)
    # lanjutkan proses utk partisi kiri dan kanan secara rekursif
    if leftMostIndex < right:
        quickSort(leftMostIndex,right,A)
    if left+1 < rightMostIndex:
        quickSort(left+1,rightMostIndex,A)
    return

def main():
    A = isiArray()
    quickSort(0, len(A)-1, A)

if __name__ == '__main__':
    main()
```

# Merge Sort

```python
def isiArray():
    nilai_n = int(input("Nilai N: "))
    A = nilai_n *[None]
    for i in range(0,nilai_n,1):
        A[i] = int(input("Nilai x: "))
    print("Kondisi Awal Array:")
    printArray(A)
    print("Proses Merge Sort:")
    return A,nilai_n

def printArray(arr):
    print("", end="")
    for i in arr:
        if (i != None):
            print(i, end=" ")
    print()

def mergesort(first,last, nilai_n, A):
    if( first < last ):
        mid = (first + last)//2
        # Urutkan bagian-1 dari array
        mergesort(first, mid, nilai_n, A)
        # Urutkan bagian-2 dari array
        mergesort(mid+1, last, nilai_n, A)
        # Merge kedua bagian
        merge(first, mid, last, nilai_n, A)

def merge(first,mid,last,nilai_n,A):
    #tempArray : array temporer utk tampung hasil merge
    #Merge bagian-1 (indexA) dan bagian-2 (indexB)
    tempArray = [None]*(nilai_n+1)
    indexA = first
    lastA = mid
    indexB = mid+1
    lastB = last
    indexC = indexA
    while( indexA <= lastA and indexB <= lastB ):
        #debugging liat ini tadi: print("perbandingan: ",A[indexA],A[indexB],A[indexC]), bakal keliatan ada ngebandingin
        #none (index a) dengan angka lain (index b)
        if( A[indexA] < A[indexB] ):
            tempArray[indexC] = A[indexA]
            indexA = indexA + 1
        else:
            tempArray[indexC] = A[indexB]
            indexB = indexB + 1
        #ini yang diperbaiki, karna index C harusnya inkremental buat kedua kasusnya, sebelumnya hanya yg else 
        indexC = indexC + 1
            #while selesai
    #Copy sisa array ke tempArray
    #Hanya satu while saja yang dikerjakan
    while (indexA <= lastA):
        tempArray[indexC] = A[indexA]
        indexA = indexA + 1
        indexC = indexC + 1
    while (indexB <= lastB ):
        tempArray[indexC] = A[indexB]
        indexB = indexB + 1
        indexC = indexC + 1
        #Salin kembali tempArray ke array semula
    printArray(A)
    indexC = first
    while (indexC <= last):
        A[indexC] = tempArray[indexC]
        indexC = indexC + 1

def main():
    A,nilai_n = isiArray()
    mergesort(0, len(A)-1, nilai_n, A)
    mid = (0 + len(A)-1)//2
    merge(0, mid, len(A)-1, nilai_n,A)
    

if __name__ == '__main__':
    main()
```
